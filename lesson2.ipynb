```python
# Write a function that accepts 2 numbers and returns the square of the sum of both numbers

def square_sum(x, y):
    sum_num = x + y
    return sum_num ** 2
square_sum(4, 5)
```

```python
import mysql.connector
from mysql.connector import Error
```

```python
def connect_db(host_name, user_name, db_name, password):
    connection = None
    try:
        connection = mysql.connector.connect(
            host = host_name,
            user = user_name,
            database = db_name,
            passwd = password
        )
    except Error as err:
        print(f'Error: {err}')
    return connection

# Define a read query function

def read_query(connection, query):
    result = None
    cursor = connection.cursor()

    try:
        cursor.execute(query)
        result = cursor.fetchall()
    except Error as err:
        print(f'Error: {err}')
    return result
```

```python
connection = connect_db(host_name='localhost', user_name='root',
                        db_name='northwind', password='1234')

query1 = '''

select orderid, freight, shipcountry
from orders
limit 100;

'''

data = read_query(connection=connection, query=query1)

data

```

```python
# convert to spreadsheet

data_list = []
for row in data:
    row = list(row)
    data_list.append(row)

orders_df = pd.DataFrame(data = data_list, columns=['Orderid', 'Freight', 'Country'])

orders_df
```
