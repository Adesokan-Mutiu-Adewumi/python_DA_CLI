## Methods for Data Wrangling in Pandas and Numpy

```python
# .head method

rent_df = pd.read_csv('House_Rent_Dataset.csv')
rent_df.head() # this will return first 5 rows in default
rent_df.head(3) # will return the first 3




```

```python
# .tail method
rent_df.tail() # last 5 by default
rent_df.tail(3) # last 3
```

```python
# .shape attribute - It returns the dimension of the data
rent_df.shape 
```

```python
# .info method - It returns qualitative characteristcis of the dataset
rent_df.info()
```

```python
# extracting patterns with .extract method and regex
pattern = r'(.*) out of (.*)'
rent_df[['house_floor', 'no_floor']] = rent_df['Floor'].str.extract(pattern)

rent_df.head(3)
```

```python
# .drop method - to remove coloum from data set

rent_df.drop(columns=['Floor'], inplace= True)
```

```python
# .colomns attribute = to return the available columns in the dataset

rent_df.columns

```

```python
# inspect column content with .unique method
rent_df['house_floor'].unique()
```

```python
# inspecting missing values with .isnull.sum

rent_df.isnull().sum()
```

```python
# handling missing values with dropna or fillna method

rent_df.dropna(inplace=True) # this will permanently delete the missing value
rent_df.head(3)
```

```python
rent_df.info()
```

```python
# .reset_index

rent_df.reset_index(drop= True, inplace= True)
```
